# Dithering and Finding the sweet spot

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import rcParams

# general plotting options 
rcParams.update({'figure.autolayout': True})
my_cmap = plt.cm.jet
my_cmap.set_under('white',1)
my_cmap.set_over('white', 300)
my_cmap.set_bad('white')
```

Data used here assumes a gaussian distribution with 1 arcsec sigma for focal x, focal y, boresight altitude and boresight azimuth. Using these distributions, we will try to find the sweet spot if there was indeed an offset during survey

```python
# load the data
data = np.load("../../data/results.npz")
x_offset_data  = data["sources"][0]['x_offsets']
y_offset_data  = data["sources"][0]['y_offsets']
xy_offset_data = data["sources"][0]['y_offsets']
az_offset_data   = data["sources"][0]['az_offsets']
alt_offset_data  = data["sources"][0]['alt_offsets']
bore_offset_data= data["sources"][0]['bore_offsets']
```

These are simple plots showing the distributions of the parameters

```python
plt.figure(1)

# x offsets
# these values are calculated using the platescale
plt.subplot(221)
plt.hist(x_offset_data['x_offsets'], bins=100)
plt.xlabel("x offsets [um]")

# y offsets
# these values are calculated using the platescale
plt.subplot(222)
plt.hist(y_offset_data['y_offsets'], bins=100)
plt.xlabel("y offsets [um]")

# boresight altitude offsets
plt.subplot(223)
plt.hist(alt_offset_data['alt_offsets'], bins=100)
plt.xlabel("Boresight altitude [arcsec]")

# boresight azimuth offsets
plt.subplot(224)
plt.hist(az_offset_data['az_offsets'], bins=100)
plt.xlabel("Boresight azimuth [arcsec]")

plt.show()
```

```python
plt.figure(1)

# x offsets
# these values are calculated using the platescale
plt.subplot(221)
plt.hist2d(x_offset_data['x_offsets'], x_offset_data['results_b'], bins=100, vmin=1, cmap=my_cmap)
plt.xlim(-100, 100)
plt.colorbar()
plt.xlabel("x offsets [um]")
plt.ylabel("SNR (b band)")

# y offsets
# these values are calculated using the platescale
plt.subplot(222)
plt.hist2d(y_offset_data['y_offsets'], y_offset_data['results_b'], bins=100, vmin=1, cmap=my_cmap)
plt.xlim(-100, 100)
plt.colorbar()
plt.xlabel("y offsets [um]")
plt.ylabel("SNR (b band)")

# boresight altitude offsets
plt.subplot(223)
plt.hist2d(alt_offset_data['alt_offsets'], alt_offset_data['results_b'], bins=100, vmin=1, cmap=my_cmap)
plt.xlim(-1.2, 1.2)
plt.colorbar()
plt.xlabel("Altitude offsets [arcsec]")
plt.ylabel("SNR (b band)")

# boresight azimuth offsets
plt.subplot(224)
plt.hist2d(az_offset_data['az_offsets'], az_offset_data['results_b'], bins=100, vmin=1, cmap=my_cmap)
plt.xlim(-1.2, 1.2)
plt.colorbar()
plt.xlabel("Azimuth offsets [arcsec]")
plt.ylabel("SNR (b band)")

plt.show()
```
